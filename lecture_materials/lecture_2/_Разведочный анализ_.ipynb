## Анализ числовых данных

### Неграфический анализ

```python
year_of_construction = dataset['Год постройки']
```

```python
max_value = year_of_construction.max()
min_value = year_of_construction.min()
print('Максимальный год постройки: ', max_value, 'Минимальный год постройки: ', min_value)
```

```python
mean_value = year_of_construction.mean()
print('Средний год постройки: ', mean_value)
```

**Альтернативный способ - медиана**

Медиана — число, которое находится ровно в середине отсортированного массива значений

[1900, 2022, 2022, 2021, 2019] # среднее = 1996.8, медиана = 2021

```python
median_value = year_of_construction.median()
print('Медиана года постройки: ', median_value)
```

**Процентили**

Процентиль — мера, в которой определенный процент всех значений в датасете равен этой мере или меньше ее

```python
percentile_10_value = year_of_construction.quantile(0.10)
percentile_25_value = year_of_construction.quantile(0.25)
percentile_50_value = year_of_construction.quantile(0.50)
percentile_75_value = year_of_construction.quantile(0.75)
percentile_90_value = year_of_construction.quantile(0.90)
print('10-й процентиль года постройки: ', percentile_10_value)
print('25-й процентиль года постройки: ', percentile_25_value)
print('50-й процентиль года постройки: ', percentile_50_value)
print('75-й процентиль года постройки: ', percentile_75_value)
print('90-й процентиль года постройки: ', percentile_90_value)
```

10% зданий построены до 1962 года.

25% зданий построены до 1979 года.

50% зданий построены до 2018 года (медиана).

# Фильтрация
Фильтрация — выделение части датасета по некоторому условию

### Кейс 1. Мы ищем квартиру в которую можно проинвестировать на этапе котлована


> фильтр = датасет[название_фильтруемой_колонки] == значение

> новый_датасет = датасет[фильтр]

```python
filter = dataset['Этап строительства']=='Котлован'
filtered_dataset = dataset[filter]

filtered_dataset.head(2)
```

### Кейс 2. Мы рассматриваем только квартиры в новостройках

> фильтр = датасет[название_фильтруемой_колонки] > значение

> новый_датасет = датасет[фильтр]

```python
filter = dataset['Год постройки']>2010
filtered_dataset = dataset[filter]

filtered_dataset.head(2)
```

### Кейс 3. Мы рассматриваем только квартиры в определенных районах.

> список_значений = [значние1, значение2]

> фильтр = датасет[название_фильтруемой_колонки].isin(список_значений)

> новый_датасет = датасет[фильтр]

```python
districts = ['Строгино', 'Щукино']

filter = dataset['Район'].isin(districts)

filtered_dataset = dataset[filter]

filtered_dataset.head()
```

### Кейс 4. Мы рассматриваем только квартиры в определенных районах, при этом нас интересуют только новостройки


> сложный_фильтр = (фильтр1) & (фильтр2)

> новый_датасет = датасет[сложный_фильтр]

```python
filter = (dataset['Район'].isin(districts)) & (dataset['Год постройки']>2010)

filtered_dataset = dataset[filter]
filtered_dataset.head()
```

Альтернативный вариант - применить нужные фильтры последовательно

```python
filtered_dataset = dataset[dataset['Район'].isin(districts)]

double_filtered_dataset = filtered_dataset[filtered_dataset['Год постройки']>2010]

double_filtered_dataset.head()
```

### Кейс 5. Мы рассматриваем все квартиры, но определенные районы нам не подходят

```python
filter = ~(dataset['Район'].isin(districts))

filtered_dataset = dataset[filter]
filtered_dataset.head()
```
